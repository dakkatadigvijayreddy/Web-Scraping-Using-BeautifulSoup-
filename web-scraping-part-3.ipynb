```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Request page source from url.
url = "https://www.scrapethissite.com/pages/simple/"
page = requests.get(url)
page.content
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify)
```

```python
# Scrap Country Names.
country_names = soup.find_all('h3',class_='country-name')
country_names
```

```python
country_name = []
for country in country_names:
    country = country.get_text(strip=True).replace("\n","")
    country_name.append(country)
country_name[:5]
```

```python
# Scrap Country Capital City.
capital_city = soup.find_all('span',class_ = "country-capital")
capital_city
```

```python
country_capital = []
for city in capital_city:
    city = city.get_text(strip=True).replace("\n","")
    country_capital.append(city)
country_capital[:5]
```

```python
# Scrap Population.
population  = soup.find_all('span',class_ = "country-population")
population
```

```python
country_population = []
for pop in population:
    pop = pop.get_text(strip=True).replace("\n","")
    country_population.append(pop)
country_population[:5]
```

```python
# Scrap Country Area
area = soup.find_all('span',class_ = "country-area")
area
```

```python
country_area = []
for ara in area:
    ara = ara.get_text(strip=True).replace("\n","")
    country_area.append(ara)
country_area[:5]
```

```python
# Store The Scraped Data.
data = pd.DataFrame()
data['country_name'] = country_name
data['country_capital'] = country_capital
data['country_population'] = country_population
data['country_area'] = country_area
data
```

```python
data.info()
```

```python
data['country_population'] = data['country_population'].astype(int)
data['country_area'] = data['country_area'].astype(float)
```

```python
data.info()
```

```python
data.to_csv("Countries Of The World.csv")
```
